```python
import numpy as np

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n


gr1 = np.mean(img,axis=2)
gr1 = np.gradient(gr1)
gr1 = np.stack(gr1, axis=2)
gr1_x = moving_average(gr1[:,:,0])
gr1_y = moving_average(gr1[:,:,1])
plt.quiver(gr1[:,:,0], gr1[:,:,1])
```

```python
def process_window(img):
    n = img.shape[0]
    def gind(p):
        return ((n-1)/2+p[0],(n-1)/2+p[1])
    vectors = np.zeros( ((n+1)*(n-1)//2, 2) )
    weights = np.zeros( (n+1)*(n-1)//2 )
    counter = 0

    def get_scale(p):
        i = p[0]
        j = p[1]
        mean = 0
        for ind in (gind((0,0)), gind((i,j)), gind((-i,j))):
            mean += img[ind]
        mean /= 3
        var = 0
        for ind in (gind((0,0)), gind((i,j)), gind((-i,j))):
            var = (mean - img[ind])**2
        var /= 2
        var = var**0.5
        w1 = 1/(2*np.linalg.norm((i,j)))
        w2 = ((1/3)**0.5 - var)
        return w1*w2
        
    for i in range(-(n-1)//2,(n-1)//2+1):
        for j in range(1, (n-1)//2+1):
            vectors[counter] = (i,j)
            weights[counter] = get_scale((i,j))
            counter += 1
    for i in range((n-1)//2):
        vectors[counter] = (i,0)
        weights[counter] = get_scale((i,0))
    return np.dot(vectors, weights)        
```
